### Identifying which schools have the best maths results (at least 80%) from the average_math column

```python
best_math_schools = schools[(schools["average_math"] / 800) * 100 >= 80][["school_name", "average_math"]].sort_values("average_math", ascending = False)
best_math_schools
```

### What are the top ten performing schools based on the combined SAT scores?


```python
top_schools = schools.copy()
top_schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top_schools = top_schools.sort_values("total_SAT", ascending = False)
```

```python
top_schools.head()
```

```python
top_10_schools = top_schools.loc[:10, ["school_name", "total_SAT"]].head(10)
print(len(top_10_schools))
```

### Which single borough has the largest standard deviation in the combined SAT score?

```python
# Identifying the several categories of borough
top_schools.borough.value_counts()
```

```python

group_borough = top_schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
group_borough
```

```python
pd.DataFrame.rename?
```

```python
group_borough.rename(columns = {"count" : "num_schools", "mean": "average_SAT", "std": "std_SAT"}, inplace = True)
group_borough
```

```python
# Filtering for the single borough with the largest standard deviation
largest_std_dev = group_borough[group_borough["std_SAT"]==group_borough["std_SAT"].max()]
largest_std_dev
```
